```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('/kaggle/input/sonar-data/sonar_data.csv',header=None)
data
```

```python
x=data.drop(columns=60,axis=1)
y=data[60]
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)

print("\nx-train\n",x_train)
print("\nx-test\n",x_test)
print("\ny-train\n",y_train)
print("\ny-test\n",x_test)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
training_accuracy=accuracy_score(x_train_prediction,y_train)
print("Accuracy on Training Data :\t",training_accuracy)

print("The Accuracy Score of Training Data is : ",(training_accuracy)*100 ,"%")
```

```python
x_test_prediction=model.predict(x_test)
testing_accuracy=accuracy_score(x_test_prediction,y_test)
print("Accuracy on Training Data :\t",testing_accuracy)

print("The Accuracy Score of Testing Data is : ",(testing_accuracy)*100 , "%")
```
