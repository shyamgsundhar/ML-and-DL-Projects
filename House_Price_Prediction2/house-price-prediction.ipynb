```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import datasets
from xgboost import XGBRegressor
from sklearn.metrics import accuracy_score,f1_score,recall_score,r2_score,recall_score,precision_score,mean_absolute_error
```

```python
datas = datasets.fetch_california_housing()
datas
```

```python
data=pd.DataFrame(datas.data,columns=datas.feature_names)
data
```

```python
data.head()
```

```python
datas
```

```python
data['Price']=datas.target
data
```

```python
print(data.shape)
print('\n')
print(data.isnull().sum())
print('\n')
```

```python
data.describe()
```

```python
correlation = data.corr()
correlation
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,fmt='.2f',annot=True,annot_kws={'size':12},cmap='Greens')
```

```python
x=data.drop(columns='Price')
y=data['Price']
print(x)
print(y)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=3)

print(xtrain)
print(xtest)
print(ytrain)
print(ytest)
```

```python
model=XGBRegressor()
model.fit(xtrain,ytrain)
```

```python
trainpredict=model.predict(xtrain)
print(trainpredict)
```

```python
xtrain
```

```python
trainpredict
```

```python
r2score=r2_score(ytrain,trainpredict)
mean=mean_absolute_error(ytrain,trainpredict)
print(r2score)
print(mean)
```
