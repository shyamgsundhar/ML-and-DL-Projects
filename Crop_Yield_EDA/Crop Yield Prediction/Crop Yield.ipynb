# Data Visualization

```python
plt.figure(figsize=(6,6))
seg=df['Season'].value_counts()
seglbl=df['Season'].unique()
plt.pie(
seg,labels=seglbl,
    autopct='%1.1f%%',
    explode = (0.06,0.05,0.05,0.07,0.08,0.05),
)
plt.show()
```

```python
sns.catplot(data=df,x="Season",aspect=3,kind='count')
```

```python
sns.catplot(data=df,x="Crop",aspect=8,kind='count')
```
