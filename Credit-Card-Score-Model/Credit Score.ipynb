```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/train.csv')
df.head()
```

```python
df.drop(["ID","Customer_ID","Name","SSN","Type_of_Loan"],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(columns='Payment_Behaviour',inplace=True)
df.head()
```

```python
categorical_var = [feature for feature in df.columns if df[feature].dtypes == 'O']
numerical_var = [feature for feature in df.columns if feature not in categorical_var]
```

```python
for feature in numerical_var:
    print(f"{feature}:")
    print(f"Number of unique values in the {feature}: {df[feature].nunique()}")
    print(f"Unique values: {df[feature].unique()}")
    print('\n')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Monthly_Inhand_Salary'].fillna(df['Monthly_Inhand_Salary'].mean(),inplace=True)
df['Num_of_Delayed_Payment']=pd.to_numeric(df['Num_of_Delayed_Payment'],errors='coerce')
df['Num_of_Delayed_Payment'].fillna(df['Num_of_Delayed_Payment'].mean(),inplace=True)
df['Amount_invested_monthly']=pd.to_numeric(df['Amount_invested_monthly'],errors='coerce')
df['Amount_invested_monthly'].fillna(df['Amount_invested_monthly'].mean(),inplace=True)
df['Num_Credit_Inquiries'].fillna(df['Num_Credit_Inquiries'].mean(),inplace=True)
df['Monthly_Balance']=pd.to_numeric(df['Monthly_Balance'],errors='coerce')
df['Monthly_Balance'].fillna(df['Monthly_Balance'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Credit_History_Age'] = df['Credit_History_Age'].str.extract(r'(\d+)')
df['Credit_History_Age']=pd.to_numeric(df['Credit_History_Age'],errors='coerce')
df['Credit_History_Age'].fillna(df['Credit_History_Age'].mean(),inplace=True)
```

```python
df.drop(df[df["Occupation"]=='_______'].index,inplace=True)
df.drop(df[df["Credit_Mix"]=='_'].index,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df['Age'] = df['Age'].replace('-500',np.nan)
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df['Age'] = df['Age'].apply(lambda x: x if x<113 else np.nan)
age_mean=df["Age"].mean()
df["Age"].fillna(age_mean, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for feature in categorical_var:
    df[feature]=le.fit_transform(df[feature])
```

```python
df['Delay_from_due_date']=df['Delay_from_due_date'].abs()
df.head()
```

```python
corr=df.corr()
plt.figure(figsize=(20,15))
sns.heatmap(corr,annot=True,fmt=".2f",cmap="coolwarm")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

x=df.drop('Credit_Score',axis=1)
y=df['Credit_Score']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier()
}
acc=[]
for model_name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    acc.append(accuracy)
    print(f"{model_name} Accuracy: {accuracy}")
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=list(models.keys()),y=acc)
plt.show()
```
